```python
from glob import glob
```

```python
data = {}
for g in glob("flags/flag*"):
    with open(g) as f:
        f.readline()
        flag = int(f.readline().strip(),16)
        
    data[float(g.split("_")[-1])] = flag
   
```

```python
data = {}
for g in glob("manyflags/flag*"):
    #print(g)
    with open(g) as f:
        data[float(g.split("_")[-1])] = []
        for line in f:
            #print("line",line)
            p = line.strip().split(",")
            #print(p)
            try:
                a = {"port": int(p[1][:-1]),
                "seq": int(p[2]),
                "stream": hex(int(p[3])),
                "sub": int(p[4]),
                "len": int(p[5]),
                "flag": int(p[6])}
            except:
                print(p)
            data[float(g.split("_")[-1])].append(a)
    #break
#ata
```
