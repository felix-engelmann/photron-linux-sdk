```python
def bytes_to_ip(i):
    return(".".join([str(b) for b in i]))

def int_to_ip(i):
    return bytes_to_ip(struct.pack(">I", i))

def ip_to_bytes(ip):
    return struct.pack(">BBBB", *map(int,ip.split(".")))

def ip_to_int(ip):
    return struct.unpack(">I", ip_to_bytes(ip))[0]
```

```python
def diff(a,b):
    for i,(ab,bb) in enumerate(zip(a,b)):
        if ab == bb:
            print(f"{bb:02x}", end="")
        else:
            print(f"\x1b[1m{bb:02x}\x1b[m", end="")
        if i%4==3:
            print(" ", end="")
    print("")
diff(b'12345678',b'12355678')
```

```python
def create_response(command, address, index, val1, val2):
    rcommand = 1
    raddress = address
    rindex, ret1, ret2 = index,0,0
    if command == 0: # get
        if (address,index, val1, val2) in responses:
            #print("reply with", responses[(address,index, val1, val2) ])
            rcommand = 1 # got
            raddress = address
            rindex, ret1, ret2 = responses[(address, index, val1, val2)]
        if address == 0x40:
            if val1 > 1:
                rcommand = 1 # got
                raddress = address
                rindex, ret1, ret2 = index, 0,0
        elif address == 0x14:
            #print("ips", hex(val1), hex(val2))
            if val1 == 0xf0000000:
                ret1 = val1
                ret2 = ip_to_int("192.168.2.63")
            elif val1 == 0xf1000000:
                ret1 = val1
                ret2 = ip_to_int("255.255.255.0")
            elif val1 == 0xf2000000:
                ret1 = val1
                ret2 = ip_to_int("0.0.0.0")
    elif command == 2:
        #b'P\xb1\x00\x00\x00\x00\x00\x00\x00\x00\x00\x00'
        rcommand = 5
        raddress = 177
        ret1=0
        ret2=0
        rindex=0
    elif command == 6:
        rcommand = 7 
        ret1 = 5637354 
    elif command == 4:
        rcommand = 5
        rindex = 1 
    r0 = rcommand << 4
    r0 += ((raddress & 0xF00) >> 8)
    r1 = raddress & 0xFF
    reply = struct.pack(">BBHII", r0,r1,rindex, ret1, ret2)
    return reply
```
