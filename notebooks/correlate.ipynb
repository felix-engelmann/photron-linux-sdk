```python
def parsedat(filename):
    vals = {}
    with open(filename, "r") as f:
        for line in f:
            if not line.startswith("PDC_Get"):
                continue
            fname = line.split(" ")[0]
            if "LIST:" in line:
                value = list(map(int,line.strip().split("LIST:")[1].split(";")[:-1]))
            else:
                parts = line.strip().split(" ; ")[1:]
                value = {k:int(v) for k,v in map(lambda x: x.split(","), parts)}
            vals[fname] = value
    return vals
```

```python
base = parsedat("impl/dump/1.dat")
base["PDC_GetColorType"]["pMode"] = 0
```

```python
for i in range(1,219):
    with open(f"impl/dump/{i}.pkl", "rb") as f:
        params = pickle.load(f)
        for k in params:
            if params[k] != og_responses[k]:
                print(k, params[k],og_responses[k])
    try:
        data = parsedat(f"impl/dump/{i}.dat")
        for k in data:
                if data[k] != base[k]:
                    print(k)
                    print(data[k])
                    print(base[k])
    except:
        print("failed")
    print("---")
```
